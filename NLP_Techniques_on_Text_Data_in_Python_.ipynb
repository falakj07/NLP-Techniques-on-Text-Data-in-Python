```python
#1. Tokenization
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize
text = "Natural language processing is interesting"
tokens = word_tokenize(text)
print(tokens)
```

```python
# Sentence Tokenization
from nltk.tokenize import sent_tokenize
text = "Python is easy . Interesting to learn."
sentences = sent_tokenize(text)
print(sentences)
```

```python
# Stop word removal
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
print(filtered_tokens)
```

```python
# Stemming
# Stemming reduces words to their root form
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
words = ["happy","crying","laughing"]
for w in words:
    print(stemmer.stem(w))
```

```python
# Lemmatization
# Lemmatization converts words to their dictionary form .

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

print(lemmatizer.lemmatize("crying"))
print(lemmatizer.lemmatize("laughing"))
print(lemmatizer.lemmatize("shouting"))
```

```python
# Part - of - speech (POS)Tagging
# POS Tagging identifies grammatical roles of words.

import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from nltk import pos_tag
from nltk.tokenize import word_tokenize

words = word_tokenize("Python is a powerfulprogramming language ")
pos_tags =pos_tag(words)
print(pos_tags)
```
